2. What is an interpreter in Python ?

   - An interpreter in Python is a software program that reads and executes Python code directly, line by line. Unlike a compiler, which translates the entire program into machine code before execution, the Python interpreter translates and runs the code on the fly. This process is what makes Python an interpreted language.

3. What are pre-defined keywords in Python ?

These keywords are part of Python's syntax, and they cannot be used as identifiers (e.g., variable, function, or class names) because they are reserved for programming constructs.



4.Can keywords be used as variable names ?

10. What is the difference between implicit and explicit type casting?
   - Type casting in Python involves converting one data type to another. It can be categorized into two types: Implicit and Explicit. Here's a detailed comparison:

Feature	Implicit Type Casting	Explicit Type Casting
Definition	Automatically performed by Python when it is safe.	Performed manually by the programmer using built-in functions.
Control	Python controls the type conversion.	The programmer has full control over the conversion.
Risk of Data Loss	No risk; Python ensures safe conversions.	Risk of data loss if conversion is incompatible or truncates data.
Use Case	Applied when types are compatible (e.g., int to float).	Used to force conversion between different types (e.g., str to int).
Error Handling	Python handles errors and avoids unsafe conversions.	Programmer must ensure the conversion is valid to avoid runtime errors.
Examples	Integer to Float, Float to Complex, etc.	String to Integer, Float to Integer, etc.

13. What is the difference between for and while loops?
  - Both for and while loops in Python are used for iteration, allowing you to repeatedly execute a block of code. However, they differ in how they work and when they are used.

Feature	for Loop	while Loop
Control Mechanism	Iterates over a sequence (e.g., list, range, string).	Repeats as long as a condition evaluates to True.
Use Case	When the number of iterations is known or the iteration is over a fixed sequence.	When the number of iterations is not predetermined, and you rely on a condition to terminate the loop.
Syntax	for variable in sequence:	while condition:
Termination	Stops when the sequence is fully traversed.	Stops when the condition becomes False.
Suitability	Best for iterating over items in a collection or range.	Best for indefinite loops or loops controlled by user input or calculations.
Common Use Cases	Iterating over lists, strings, dictionaries, or ranges.	Repeatedly prompting user input, waiting for specific conditions.
Risk of Infinite Loop	Low, as it depends on a sequence.	Higher, if the condition is never met or improperly updated.

#6. Write a function to demonstrate the behavior of mutable and immutable argument

```python
def modify_values(immutable_arg, mutable_arg):
    # Attempting to modify immutable argument (this won't change the original value)
    print("Before modifying immutable_arg:", immutable_arg)
    immutable_arg = 100  # This won't change the original value outside the function
    print("After modifying immutable_arg:", immutable_arg)

    # Modifying mutable argument (this will change the original value outside the function)
    print("\nBefore modifying mutable_arg:", mutable_arg)
    mutable_arg[0] = 999  # This changes the first element of the list
    print("After modifying mutable_arg:", mutable_arg)

# Testing the function
immutable_value = 10  # Immutable argument (integer)
mutable_value = [1, 2, 3]  # Mutable argument (list)

# Call the function with both types of arguments
modify_values(immutable_value, mutable_value)

# Show the original values after function call
print("\nOriginal immutable_value after function call:", immutable_value)
print("Original mutable_value after function call:", mutable_value)
```

#8. Write a program to demonstrate the use of logical operators.

```python
# Define some variables for demonstration
x = 10
y = 5
z = 0

# Using 'and' operator
print("Using 'and' operator:")
print(f"x > y and y > z: {x > y and y > z}")  # True and True -> True
print(f"x < y and y > z: {x < y and y > z}")  # False and True -> False
print()

# Using 'or' operator
print("Using 'or' operator:")
print(f"x > y or y > z: {x > y or y > z}")  # True or True -> True
print(f"x < y or y > z: {x < y or y > z}")  # False or True -> True
print(f"x < y or z > y: {x < y or z > y}")  # False or False -> False
print()

# Using 'not' operator
print("Using 'not' operator:")
print(f"not (x < y): {not (x < y)}")  # not False -> True
print(f"not (y > z): {not (y > z)}")  # not True -> False
print(f"not (x == 10): {not (x == 10)}")  # not True -> False
print(f"not (z == 0): {not (z == 0)}")  # not True -> False
```

#9.Write a Python program to convert user input from string to integer, float, and boolean types.

```python
# Taking user input as a string
user_input = input("Enter a value: ")

# Convert the input to integer
try:
    int_value = int(user_input)
    print(f"Integer value: {int_value}")
except ValueError:
    print("The input cannot be converted to an integer.")

# Convert the input to float
try:
    float_value = float(user_input)
    print(f"Float value: {float_value}")
except ValueError:
    print("The input cannot be converted to a float.")

# Convert the input to boolean
# In Python, empty strings are considered False, and any non-empty string is considered True.
bool_value = bool(user_input)
print(f"Boolean value: {bool_value}")
```

#10.Write code to demonstrate type casting with list elements

```python
# Original list with elements of different types
my_list = ['10', '20.5', 'True', '100']

# Convert string elements to integers and floats, and strings to booleans
# Example of type casting with list elements

# Create an empty list to store the converted values
converted_list = []

# Convert each element to an integer (if possible), float, and boolean
for item in my_list:
    try:
        # Convert to integer (if possible)
        int_value = int(item)
        converted_list.append(('int', item, int_value))
    except ValueError:
        # If cannot convert to integer, try converting to float
        try:
            float_value = float(item)
            converted_list.append(('float', item, float_value))
        except ValueError:
            # If cannot convert to float, try converting to boolean
            bool_value = bool(item)
            converted_list.append(('bool', item, bool_value))

# Print the results
for data_type, original_value, converted_value in converted_list:
    print(f"Original value: {original_value} -> Type: {data_type} -> Converted value: {converted_value}")
```

#11.  Write a program that checks if a number is positive, negative, or zero

```python
# Function to check if a number is positive, negative, or zero
def check_number(number):
    if number > 0:
        return "The number is positive."
    elif number < 0:
        return "The number is negative."
    else:
        return "The number is zero."

# Taking user input
try:
    num = float(input("Enter a number: "))
    result = check_number(num)
    print(result)
except ValueError:
    print("Invalid input! Please enter a valid number.")
```

#13.  Write a Python program to find the sum of all even numbers between 1 and 50

```python
# Initialize the sum variable
even_sum = 0

# Loop through numbers from 1 to 50
for num in range(1, 51):
    if num % 2 == 0:  # Check if the number is even
        even_sum += num  # Add the even number to the sum

# Print the sum of all even numbers
print(f"The sum of all even numbers between 1 and 50 is: {even_sum}")
```

#14.Write a program to reverse a string using a while loop

```python
# Function to reverse a string using a while loop
def reverse_string(input_string):
    reversed_string = ""  # Initialize an empty string to store the reversed string
    index = len(input_string) - 1  # Set index to the last character of the input string

    while index >= 0:
        reversed_string += input_string[index]  # Add the character to the reversed string
        index -= 1  # Move to the previous character

    return reversed_string

# Taking user input
user_input = input("Enter a string to reverse: ")

# Reversing the string and printing the result
reversed_str = reverse_string(user_input)
print(f"Reversed string: {reversed_str}")
```

#15.Write a Python program to calculate the factorial of a number provided by the user using a while loop

```python
# Function to calculate factorial using a while loop
def factorial(n):
    result = 1  # Initialize result to 1 (as factorial starts from 1)

    while n > 1:  # Loop continues until n becomes 1
        result *= n  # Multiply result by current value of n
        n -= 1  # Decrease n by 1

    return result

# Taking user input
try:
    num = int(input("Enter a number to calculate its factorial: "))

    # Check for negative number
    if num < 0:
        print("Factorial does not exist for negative numbers.")
    else:
        # Calculate and print the factorial
        fact = factorial(num)
        print(f"The factorial of {num} is: {fact}")

except ValueError:
    print("Invalid input! Please enter a valid number.")
```
